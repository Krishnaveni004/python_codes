```python
# Import the Pandas library
import pandas as pd
```

```python
# Load the train dataset to create a DataFrame
train_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/train.csv"
train = pd.read_csv(train_url)
```

```python
#Load the test dataset to create a DataFrame
test_url = "http://s3.amazonaws.com/assets.datacamp.com/course/Kaggle/test.csv"
test=pd.read_csv(test_url)
```

```python
#Print the `head` of the train and test dataframes
print(train.head())
print(test.head())
```

```python
# Females that survived vs Females that passed away
print(train["Survived"][train["Sex"]=='female'].value_counts())
```

```python
#Next we check for age dependency, but before that we see that age count is only 714
train.describe()
```

```python
#Fill up missing data
train["Age"] = train["Age"].fillna(train["Age"].median())
```

```python
#Check age count again
train.describe()
```

```python
# Create the column Child and assign to 'NaN'
train["Child"] = float('NaN')
```

```python
# Assign 1 to passengers under 18, 0 to those 18 or older. Print the new column.
train["Child"][train["Age"]<18]=1
train["Child"][train["Age"]>=18]=0
print(train["Child"])
```

```python
# Print normalized Survival Rates for passengers under 18
print(train["Survived"][train["Child"] == 1].value_counts(normalize = True))

# Print normalized Survival Rates for passengers 18 or older
print(train["Survived"][train["Child"] == 0].value_counts(normalize=True))
```

```python
# Create a copy of test: test_one for making a trial_basic prediction
test_one=test

# Initialize a Survived column to 0
test_one["Survived"]=0

# Set Survived to 1 if Sex equals "female" and print the `Survived` column from `test_one`
test_one["Survived"][test_one["Sex"]=="female"]=1
print(test_one["Survived"])
```

```python
#####Decision Tree#####
# Import the Numpy library
import numpy as np

# Import 'tree' from scikit-learn library
from sklearn import tree
```

```python
# Convert the male and female groups to integer form
train["Sex"][train["Sex"] == "male"] = 0
train["Sex"][train["Sex"]=="female"]=1
```

```python
# Impute the Embarked variable
train["Embarked"] = train["Embarked"].fillna('S')
```

```python
# Convert the Embarked classes to integer form
train["Embarked"][train["Embarked"] == "S"] = 0
train["Embarked"][train["Embarked"]=="C"] = 1
train["Embarked"][train["Embarked"]=="Q"] = 2
```

```python
#Print the Sex and Embarked columns
print(train["Sex"])
print(train["Embarked"])
```

```python
# Print the train data to see the available features
print(train)
```

```python
# Create the target and features numpy arrays: target, features_one
target = train["Survived"].values
features_one = train[["Pclass", "Sex", "Age", "Fare"]].values
```

```python
# Fit your first decision tree: my_tree_one
my_tree_one = tree.DecisionTreeClassifier()
my_tree_one = my_tree_one.fit(features_one,target)
```

```python
# Look at the importance and score of the included features
print(my_tree_one.feature_importances_)
print(my_tree_one.score(features_one, target))
```

```python
#####   START PREDICTING  #####
# Impute the missing value with the median
test.Fare[152] =  test["Fare"].median()
```

```python
#Checking features of test dataset
test.head(1)
```

```python
#Check for missing values
test.describe()
```

```python
#Age is missing...fill in
test["Age"] = test["Age"].fillna(train["Age"].median())
```

```python
#Changing to numeric forms as in training dataset

# Convert the male and female groups to integer form
test["Sex"][test["Sex"] == "male"] = 0
test["Sex"][test["Sex"]=="female"]=1

# Impute the Embarked variable
test["Embarked"] = test["Embarked"].fillna('S')

# Convert the Embarked classes to integer form
test["Embarked"][test["Embarked"] == "S"] = 0
test["Embarked"][test["Embarked"]=="C"] = 1
test["Embarked"][test["Embarked"]=="Q"] = 2
```

```python
# Extract the features from the test set: Pclass, Sex, Age, and Fare.
test_features = test[['Pclass', 'Sex', 'Age', 'Fare']].values
print(test_features)
```

```python
# Make your prediction using the test set
my_prediction = my_tree_one.predict(test_features)
print(my_prediction)
```

```python
# Create a data frame with two columns: PassengerId & Survived. Survived contains your predictions
PassengerId =np.array(test["PassengerId"]).astype(int)
my_solution = pd.DataFrame(my_prediction, PassengerId, columns = ["Survived"])
print(my_solution)
```
